```python
%%writefile app.py

import streamlit as st
import requests

# Backend API URL
API_BASE_URL = "http://127.0.0.1:8000"
 # Replace with deployed backend URL

st.set_page_config(page_title="AI Mock Interview", layout="centered")
st.title("ğŸ¤ AI-Powered Mock Interview Platform")
st.markdown("Practice interviews with real-time feedback powered by AI ğŸ’¡")

# --- Select Role ---
role = st.selectbox("Choose your interview role:", ["Data Scientist", "Software Engineer", "Product Manager"])

# --- Generate Question ---
if st.button("ğŸ¯ Generate Interview Question"):
    with st.spinner("Generating question..."):
        response = requests.post(f"{API_BASE_URL}/generate_question", json={"role": role})
        if response.status_code == 200:
            question = response.json().get("question")
            st.session_state["current_question"] = question
        else:
            st.error("Failed to fetch question. Try again.")

# Display generated question
if "current_question" in st.session_state:
    st.subheader("ğŸ“ Your Question")
    st.write(st.session_state["current_question"])

    # --- Text Input Answer ---
    user_answer = st.text_area("Enter your answer below ğŸ‘‡", height=150)

    # --- Submit Answer for Evaluation ---
    if st.button("âœ… Submit Answer"):
        if user_answer.strip() == "":
            st.warning("Please enter an answer before submitting.")
        else:
            with st.spinner("Evaluating your response..."):
                response = requests.post(f"{API_BASE_URL}/evaluate_answer", json={
                    "question": st.session_state["current_question"],
                    "answer": user_answer
                })

                if response.status_code == 200:
                    result = response.json()
                    st.success("âœ… Evaluation Complete")

                    st.markdown(f"**Semantic Score:** {result['semantic_score']} / 1.0")
                    st.markdown(f"**Feedback:** {result['feedback']}")
                    st.markdown(f"**Tone & Sentiment:** {result['sentiment']}")

                else:
                    st.error("Evaluation failed. Please try again.")

# --- Footer ---
st.markdown("---")
st.caption("Built with â¤ï¸ using Streamlit, GPT, and Sentence-BERT")
```

```python

from fastapi import FastAPI
from pydantic import BaseModel
import openai
import os
from dotenv import load_dotenv

# âœ… Load environment variables from .env
load_dotenv()

# âœ… Set the OpenAI API key from .env
openai.api_key = os.getenv("OPENAI_API_KEY")

app = FastAPI()

class Role(BaseModel):
    role: str

@app.post("/generate_question")
def generate_question(data: Role):
    prompt = f"""You are an expert interviewer. Generate 3 unique and challenging interview questions for the role of {data.role}. Include both technical and behavioral questions."""

    response = openai.ChatCompletion.create(
        model="gpt-3.5-turbo",
        messages=[
            {"role": "user", "content": prompt}
        ],
        temperature=0.7,
        max_tokens=300
    )

    return {"questions": response.choices[0].message['content'].strip()}
print("OpenAI KEY LOADED:", openai.api_key)
```
